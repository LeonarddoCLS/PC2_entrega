```python
!pip install openpyxl
import pandas as pd
link='https://github.com/LeonarddoCLS/PC2_entrega/raw/refs/heads/main/HDI%20(1).xlsx'
data=pd.read_excel(link)
data
```

```python
#Eliminamos las columnas con las que no vamos a trabajar
#Las eliminamos por posiciÃ³n, para una mayor facilidad

data_clean=data.copy()

byeColumns=[data_clean.columns[i] for i in [3,5,7,9,11,12,13,14]]
data_clean.drop(columns=byeColumns, inplace=True)
data_clean
```

```python
#Renombro las columna que voy a utilizar
#Creamos un diccionario

goodNames=['reconocer', 'Country','HDI','Life_expectancy','Expected_years_of_schooling', 'Mean_years_of_schooling', 'GNI_per_capita']
#Creamos el diccionario con nuevos nombres
{old:new for old,new in zip(data_clean.columns,goodNames)}
```

```python
#Reemplazmos los nuevos valores
data_clean.rename(columns={old:new for old, new in zip(data_clean.columns, goodNames)}, inplace=True)
data_clean
```

```python
#ahora eliminamos las filas que no vamos a utilizar
#la fila en donde empiezan los datos es a partir de la 7, eliminamos las primeras filas 6
byeRows=[0,1,2,3,4,5,6]
data_clean.drop(index=byeRows, inplace=True)
data_clean
```

```python
#Elimino las filas que no tienen informaciÃ³n completa de los paÃ­ses
data_clean_new=data_clean.copy()
#Para trabajar con regex, necesito transformar mis valores de la columna reconocer a .string
data_clean_new['reconocer'] = data_clean_new['reconocer'].astype(str)
remove_rows=data_clean_new[~data_clean_new.reconocer.str.contains(pat=r'\d',regex=True)]
#guardo la informaciÃ³n de los indices de las filas que no trabajarÃ©
remove_rows_index=list(remove_rows.index)
data_clean_new.drop(index=remove_rows_index, inplace=True)
data_clean_new
```

```python
#Elmino columna reconocer
data_clean_new.drop(columns=data_clean_new.columns[0], inplace=True)
data_clean_new
```

```python
#Reseteo indices
data_clean_new2=data_clean_new.copy()
data_clean_new2.reset_index(drop=True,inplace=True)
data_clean_new2.index=data_clean_new2.index+1
data_clean_new2
```

```python
data_clean_new2.info()
```

```python
data_clean_new2=data_clean_new2.astype(str)
byeSpaces=lambda x: x.str.strip()
#ya converitÃ­ todos mis valores a str
data_clean_new2.iloc[:,[0,5]]=data_clean_new2.iloc[:,[0,5]].apply(byeSpaces)
data_clean_new2
```

```python
#Me aseguro de que no haya digitos raros en cada columna de valores que transformarÃ© en nÃºmericos
data_clean_new2.HDI[~data_clean_new2.HDI.str.contains('^\d+.\d+')]
```

```python
data_clean_new2.Life_expectancy[~data_clean_new2.Life_expectancy.str.contains('^\d+.\d+')]
```

```python
data_clean_new2.Expected_years_of_schooling[~data_clean_new2.Expected_years_of_schooling.str.contains('^\d+.\d+')]
```

```python
data_clean_new2.Mean_years_of_schooling[~data_clean_new2.Mean_years_of_schooling.str.contains('^\d+.\d+')]
```

```python
data_clean_new2.GNI_per_capita[~data_clean_new2.GNI_per_capita.str.contains('^\d+.\d+')]
```

Ningun valor tiene coma o algÃºn otro caracter raro por equivocaciÃ³n


```python
data_clean_new2['HDI']=pd.to_numeric(data_clean_new2.HDI)
data_clean_new2['Life_expectancy']=pd.to_numeric(data_clean_new2.Life_expectancy)
data_clean_new2['Expected_years_of_schooling']=pd.to_numeric(data_clean_new2.Expected_years_of_schooling)
data_clean_new2['Mean_years_of_schooling']=pd.to_numeric(data_clean_new2.Mean_years_of_schooling)
data_clean_new2['GNI_per_capita']=pd.to_numeric(data_clean_new2.GNI_per_capita)
data_clean_new2.info()
```

```python
#Para simplicicar analisis, quitamos todos los caracteres especiales
from unidecode import unidecode

data_clean_new2['Country']=data_clean_new2.Country.apply(unidecode)
```

```python
#Colocamos en mayÃºsculas para facilitar anÃ¡lisis posterior
data_clean_new2['Country']=data_clean_new2.Country.str.upper()
data_clean_new2
```

Finalmente, creamos una categorÃ­a respecto al HDI

```python
data_clean_new2['HDI_State']=pd.cut(data_clean_new2.HDI,bins=[0,0.25,0.50,0.75,1],labels=['very_bad','bad','good','very_good'],ordered=True)
data_clean_new2
```
